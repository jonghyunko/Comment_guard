```python
import re
```

```python
def remove_special_characters(text):
    # 특수문자 및 숫자를 모두 제거
    text = re.sub('[^a-zA-Z가-힣\s]', '', text)
    return text

# "문장" 열에 적용하여 값 수정
df['문장'] = df['문장'].apply(remove_special_characters)

# 결과 확인
print(df['문장'])
```

```python
from konlpy.tag import Okt

def remove_stopwords_and_special_characters(text, stopwords, special_chars_to_keep):
    okt = Okt()
    tokens = okt.morphs(text, stem=True)  # 형태소 분석
    tokens = [word for word in tokens if word not in stopwords]  # 불용어 제거
    tokens = [re.sub(r'[^a-zA-Z가-힣\s{}]'.format(re.escape("".join(special_chars_to_keep))), '', word) for word in tokens]  # 특수문자 제거
    tokens = [word for word in tokens if word]  # 빈 문자열 제거
    return " ".join(tokens)


stopwords = ["은", "는", "이", "가", "을", "를", "으로", "에서"]
# 사용자가 유지하고 싶은 특수문자 정의
special_chars_to_keep = ['?', '!']

# 예시 문장
sentence = "불용어 제거 및 특수문자 유지 테스트용 문장입니다. 잘 될까요? 궁금하네요!"

# 불용어 제거 및 특수문자 유지 적용
result = remove_stopwords_and_special_characters(sentence, stopwords, special_chars_to_keep)
print(result)
```

```python
df['문장'] = df['문장'].apply(lambda x: remove_stopwords_and_special_characters(x, stopwords, special_chars_to_keep))
```

```python
print(df['문장'])
```

```python
# Okt 객체 생성
okt = Okt()

# 토큰화 함수 정의
def tokenize(text):
    tokens = okt.morphs(text, stem=True)
    return tokens

# '문장' 열에 토큰화 적용
df['토큰'] = df['문장'].apply(tokenize)

# 결과 확인
print(df[['문장', '토큰']])
```

```python
# 정수 인코딩
from keras.preprocessing.text import Tokenizer

# Tokenizer 객체 생성
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['토큰'])

# 각 토큰을 정수로 변환
df['정수인코딩'] = tokenizer.texts_to_sequences(df['토큰'])

# 결과 확인
print(df[['토큰', '정수인코딩']])
```

```python
from keras.preprocessing.sequence import pad_sequences

# 최대 시퀀스 길이 설정
max_sequence_length = 50

# 패딩 적용
padded_sequences = pad_sequences(df['정수인코딩'], maxlen=max_sequence_length, padding='post', truncating='post')

# 결과 확인
df['패딩된시퀀스'] = list(padded_sequences)
print(df[['토큰', '정수인코딩', '패딩된시퀀스']])
```
